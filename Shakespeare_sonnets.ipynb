```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
all_sonnets = requests.get('http://www.gutenberg.org/files/1041/1041-h/1041-h.htm')
```

```python
sonnet_page = BeautifulSoup(all_sonnets.text, 'lxml')
```

```python
sonnet_page
```

```python
sonnets = []
for tag in sonnet_page.findAll('p','poem'):
    sonnets.append(tag.get_text().strip())
print(len(sonnets))
for x in range(len(sonnets)):
    print(sonnets[x])
```

```python
titles = []
for tag in sonnet_page.findAll('h3'):
    titles.append(tag.get_text().strip())
print(len(titles))
print(titles)
```

```python
authors = []
forms = []
# raw_name = sonnet_page.find('p', 't2')
# name = raw_name.get_text().strip('by')
for i in range(len(titles)):
    authors.append('William Shakespeare')
    forms.append('Sonnet')
print(authors[0])
print(forms[0])
len(authors)
```

```python
sonnet_df = pd.DataFrame({'Author':authors, 'Title':titles, 'Poem':sonnets, 'Form':forms, 'Era':'Renaissance', 'Type':'Love'})
sonnet_df
```

```python
sonnet_df.to_csv('Shakespeare_Sonnets.csv')
```
