```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('http://www.gutenberg.org/files/1141/1141-h/1141-h.htm')
```

```python
all_poems = BeautifulSoup(page.text, 'lxml')
all_poems
```

```python
titles = []
for tag in all_poems.find_all('span','smcap'):
    titles.append(tag.get_text().strip())
#     print(tag.next_element.next_element.next_element.next_element.next_element.next_element)
print(len(titles))
print(titles)
```

```python
print(titles.index('Flower of Love'))
print(titles.index('Preface'))
titles2 = list(titles[3:22])
titles2.remove('Sonnets')
print(len(titles2))
titles2
```

```python
poems = []
poem = ''
for tag in all_poems.find_all('p','poetry'):
    tagString = str(tag.next_sibling.next_sibling)
    poem += tag.get_text() + '\n'
    if '<h3>' in tagString or '<h2>' in tagString \
    or 'Katakolo' in tagString or 'Florence' in tagString:
        poems.append(poem.strip())
        poem = ''
print(len(poems))
for poem in poems:
    print(f"{poem}\n")
    print("-"*40)
# LIBERTATIS SACRA FAMES
# PORTIA
```

```python
df = pd.DataFrame({'Author':'Oscar Wilde', 'Title':titles2, 'Poem':poems, 'Era':'Victorian'})
df
```

```python
df.to_csv('Wilde_Poems.csv')
```
