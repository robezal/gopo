### 100 Best Poems

```python
all_poems = requests.get('https://100.best-poems.net/top-100-best-poems.html')
```

```python
poems = BeautifulSoup(all_poems.text, 'lxml')
```

```python
titles_authors = []
for tag in poems.findAll('ul'):
    try:
        if 'by' in tag.get_text():
            titles_authors.append(tag.get_text().strip())
    except:
        pass
print(len(titles_authors))
titles_authors[0:12]
```

```python
titles = []
authors = []
for name in titles_authors:
    x = name.split('by', 2)
    if len(x) > 2:
        x[0] = x[0] + 'by' + x[1]
    titles.append(x[0].strip())
    authors.append(x[-1].strip())
```

```python
print(titles)
```

```python
print(authors)
```

```python
poetry = pd.DataFrame({'Author':authors, 'Title':titles})
poetry
```

```python
poem_links = []
for tag in poems.findAll('ul'):
    link = tag.find('a')['href']
    if 'https' not in link:
        poem_links.append(link)
print(len(poem_links))
poem_links[0:5]
```

```python
# for link in poem_links: 
#     poem_link = requests.get('https://100.best-poems.net' + link)
#     print(poem_link.text)
def get_poem(link):
    poem_link = 'https://100.best-poems.net' + link
    poetry = requests.get(poem_link)
    bsp = BeautifulSoup(poetry.text, 'lxml') 
    poem = bsp.find("div", "content-poetry").get_text()
    return poem.strip('\n')
```

```python
poems = []
# iterate through links
for pLink in poem_links:
    # call method to get poem
    poem = get_poem(pLink)
    # add to poem list
    poems.append(poem)
len(poems)
```

```python
for poem in poems:
    print(poem)
poems[1]
```

```python
poetry = pd.DataFrame({'Author':authors, 'Title':titles, 'Poem':poems})
poetry
```

```python
poetry.to_csv('top85_poems.csv')
```
