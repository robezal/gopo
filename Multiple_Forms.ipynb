```python
df14 = pd.read_csv('Shakespeare_Sonnets.csv', index_col = 0)
df15 = pd.read_csv('E_Browning_Sonnets.csv', index_col = 0)
df16 = pd.read_csv('ps_data_non_shakespeare_sonnets.csv')
df14.head()
```

```python
df4 = df14[['Author', 'Poem', 'Form']].copy()
df5 = df15[['Author', 'Poem', 'Form']].copy()
df6 = df16[['author', 'content', 'form']].copy()
df6.columns = ['Author', 'Poem', 'Form']
```

```python
df17 = pd.read_csv('Tibullus_Williams_Elegies.csv', index_col = 0)
df7 = df17[['Author','Content','Form']]
df7.columns = ['Author', 'Poem', 'Form']
df7.head()
```

```python
df18a = pd.read_csv('Haiku.txt', names = ['Poem', 'Author'])
df18b = df18a[['Author','Poem']]
df18 = pd.DataFrame({'Author':df18b.Author, 'Poem':df18b.Poem, 'Form':'Haiku'})
df18
```

```python
writer = []
df8 = df18.copy()
for name in df8['Author']:
    if name != float:
#         print(name.strip(' ').strip('-').strip(' '))
        writer.append(name.strip(' ').strip('-').strip(' '))
df8.Author = writer.copy()
df8.to_csv('Haiku.csv')
df8.head()
```

```python
df19 = pd.read_csv('Elegies_100.csv', index_col = 0)
df9 = df19[['Author', 'Poem', 'Form']].copy()
df9.head()
```

```python
df13 = pd.read_csv('Ballads_100.csv', index_col = 0)
df3 = df13[['Author', 'Poem', 'Form']].copy()
df3.head()
```

```python
df1 = df4.append(df5).append(df6).append(df7).append(df8).append(df9).append(df3)
df1
```
