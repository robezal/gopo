```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
all_poems = requests.get('http://www.gutenberg.org/files/1322/1322-h/1322-h.htm')
```

```python
poems_page = BeautifulSoup(all_poems.text, 'lxml')
```

```python
poems_page
```

```python
poems = []
idx = 0
for tag in poems_page.findAll('pre'):
    x = tag.get_text().replace("â\x80\x99","'")
    y = x.replace('â\x80\x9c','"')
    z = y.replace('â\x80\x9d','"')
    if tag.previous_element.previous_element.previous_element.name == 'pre':
        idx -= 1
        poems[idx] += z
    else:
        poems.append(z.strip())
    idx += 1
del poems[0]
del poems[-1]
print(len(poems))
for poem in poems:
    print(f'{poem}\n')
    print('-'*40)
```

```python
titles2 = []
for tag in poems_page.findAll('p'):
    if tag.get_text().strip() is not '':
        titles2.append(tag.get_text().strip())
print(len(titles2))
```

```python
print(titles2.index('Good-Bye My Fancy!'))
titles = list(titles2[:378])
```

```python
print(len(titles))
print(titles.index('BOOK I. INSCRIPTIONS'))
del titles[:2]
```

```python
print(titles[0])
print(len(titles))
```

```python
print(titles.index('Perfections'))
titles.insert(118, "When I Heard the Learn'd Astronomer")
print(len(titles))
```

```python
titles.insert(0, "Walt Whitman")
titles.insert(309, "FANCIES AT NAVESINK")
```

```python
titles_cleaned = []
for txt in titles:
    x = txt.replace("â\x80\x99","'")
    y = x.replace('â\x80\x9c','"')
    z = y.replace('â\x80\x9d','"')
    titles_cleaned.append(z)
print(len(titles_cleaned))
```

```python
print(titles_cleaned)
```

```python
poems_df = pd.DataFrame({'Author':'Walt Whitman','Title':titles_cleaned,'Poem':poems})
```
