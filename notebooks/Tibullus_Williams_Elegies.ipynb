```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
all_poems = requests.get('http://www.gutenberg.org/files/9610/9610-h/9610-h.htm')
```

```python
poems_page = BeautifulSoup(all_poems.text, 'lxml')
```

```python
poems_page
```

```python
poems = []
idx = 0
for tag in poems_page.findAll('pre'):
    tagString = str(tag.previous_sibling.previous_sibling)
    if 'The Sibyl' in tagString or 'To Pholoë' in tagString \
        or 'To Marathus' in tagString or "L'Envoi" in tagString:
        idx -= 1
        poems[idx] += tag.get_text()
    else:
        poems.append(tag.get_text().strip())
    idx += 1
print(len(poems))
for poem in poems:
    print(f'{poem}\n')
    print('-'*40)
```

```python
print(len(poems))
del poems[29]
print(len(poems))
del poems[:4]
```

```python
print(len(poems))
for poem in poems:
    print(f'{poem}\n')
    print('-'*40)
```

```python
titles = []
for tag in poems_page.findAll('h2'):
    tagger = tag.next_sibling.next_sibling
    if 'center' in str(tagger):
        titles.append(tagger.get_text().strip())
    if "OVID'S LAMENT" in tag.get_text():
        titles.append(tag.get_text().strip())
print(len(titles))
for title in titles:
    print(f'{title}\n')
    print('*'*42)
```

```python
df = pd.DataFrame({'Author':'Tibullus', 'Title':titles, 'Content':poems, 'Form':'Elegy', 'Era':'Modern'})
```
