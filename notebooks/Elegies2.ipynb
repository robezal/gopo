```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
from os import path
import re
```

```python
main_page = requests.get('https://www.poemhunter.com/poems/elegies/')
# second_page = requests.get('https://www.poemhunter.com/poems/elegies/page-2/')
```

```python
pages = []
pages.append(BeautifulSoup(main_page.text, 'lxml'))
# pages.append(BeautifulSoup(second_page.text, 'lxml'))
```

```python
cp = re.compile(".*cursor:pointer")
```

```python
temp = []
titles = []
authors = []
links = []
for page in pages:
    for tag in page.find_all('li', style=cp):
        temp = tag.get_text().split('\n')
        titles.append(temp[2])
        authors.append(temp[4].strip('\t').strip('\r'))
        lst = str(tag).split("'")
        links.append(lst[1])
print(len(titles))
print(titles)
print(len(authors))
print(authors)
print(len(links))
print(links)
```

```python
poems = []
for tag in page.find_all('p', style="display:block; width:330px; float:left;"):
    poem = str(tag)
    poem = poem.replace('<br/>','\n')
    poem = poem.replace('<i>', '')
    poem = poem.replace('</i>', '')
    poem = poem.replace('</p>', '')
    poem = poem.split('\n',1)[1]
    poem= poem.strip()
    poems.append(poem)
#     poems.append(tag.get_text())
[print(poem) for poem in poems]
```

```python
def get_poem(link):
    db = re.compile("display:block.*")
    poem_link = 'https://www.poemhunter.com/' + link
    poetry = requests.get(poem_link)
    bsp = BeautifulSoup(poetry.text, 'lxml')
    html_poem = bsp.find('p', style=db)
    return str(html_poem)
```

```python
poems = []
for pLink in links:
    poem = get_poem(pLink)
    poem = poem.replace('<br/>','\n')
    poem = poem.replace('<i>', '')
    poem = poem.replace('</i>', '')
    poem = poem.replace('</p>', '')
    poem = poem.split('\n',1)[1]
    poem= poem.strip()
    poems.append(poem)
len(poems)
```

```python
for poem in poems:
    print(poem)
    print('-'*45)
```

```python
df = pd.DataFrame({'Author':authors, 'Title':titles, 'Poem':poems, 'Form':'Elegy'})
df.head()
```

```python
df.to_csv('Elegies2_21.csv')
```
