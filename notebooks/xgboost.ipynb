```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
# from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('poetry_society_data_cleaned01.csv', index_col = 0)
df
```

```python
myvectorizer = CountVectorizer()
X = myvectorizer.fit_transform(df['content'])
```

```python
myvectorizer.get_feature_names()
```

```python
X.shape
```

```python
# Non-zero cells
print(X.nnz)
100*X.nnz / (X.shape[0]*X.shape[1])
```

```python
y = df.age
```

```python
# y_dummies = pd.get_dummies(y, drop_first=True)
# y_dummies
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y.values, random_state=10)
```

```python
from xgboost import XGBClassifier
my_model = XGBClassifier()
```

```python
# Add silent=True to avoid printing out updates with each cycle
my_model.fit(X_train, y_train, verbose=False)
```

```python
# 0.9243697478991597
accuracy_score(y_pred, y_test)
```

```python
my_model_2 = XGBClassifier(n_estimators=100, learning_rate=0.05)
```

```python
my_model_2.fit(X_train, y_train, verbose=False)
```

```python
y_pred2 = my_model_2.predict(X_test)
acc = accuracy_score(y_pred2, y_test)
print('Accuracy Score: ', acc)
```
