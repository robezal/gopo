```python
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
all_sonnets = requests.get('http://www.gutenberg.org/files/2002/2002-h/2002-h.htm')
```

```python
sonnet_page = BeautifulSoup(all_sonnets.text, 'lxml')
```

```python
sonnet_page
```

```python
poems = []
for tag in sonnet_page.findAll('p', 'poetry'):
    poems.append(tag.get_text().strip())
print(len(poems))
for i in range(len(poems)):
    print(f'{poems[i]}\n')
```

```python
titles = []
for tag in sonnet_page.findAll('h2'):
    titles.append(tag.get_text().strip())
titles.pop(0)
print(len(titles))
print(titles)
```

```python
sonnet_df = pd.DataFrame({'Author':'Elizabeth Barrett Browning', 'Title':titles, 'Poem':poems, 'Form':'Sonnet', 'Era':'Victorian','Type':'Love'})
```

```python
sonnet_df
```

```python
sonnet_df.to_csv('E_Browning_Sonnets.csv')
```
