```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Multiple_Forms.csv', index_col = 0)
df.head()
```

```python
df.count()
```

```python
myvectorizer = CountVectorizer()
X = myvectorizer.fit_transform(df['Poem'])
```

```python
X[0].shape
```

```python
myvectorizer.get_feature_names()
```

```python
X.shape
```

```python
#Non-zero cells
print(X.nnz)
100*X.nnz / (X.shape[0]*X.shape[1])
```

```python
y = df.Form
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=5)
```

```python
# pip install xgboost
```

```python
from xgboost import XGBClassifier
my_model = XGBClassifier()
```

```python
my_model.fit(X_train, y_train, verbose=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = my_model.predict(X_test)
```

```python
accuracy_score(y_pred, y_test)
```

```python
my_model_2 = XGBClassifier(n_estimators=25, learning_rate=0.3)
my_model_2.fit(X_train, y_train)
```

```python
y_pred2 = my_model_2.predict(X_test)
acc = accuracy_score(y_pred2, y_test)
print('Accuracy Score: ', acc)
```
